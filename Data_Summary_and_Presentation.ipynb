Stemgraphic is an easy to use python package providing a solution to these limitations (no size limit, graphical tool). It also supports categorical and text as input.

## Cummulative Frequency Distribution

Reusing the previous dataset...

```python
import numpy as np
# getting the data counts and bins from the function
count, bins_count = np.histogram(distance_values, bins=10)
# Probabilidy Density Function. Count is a 1D array.
pdf = count / sum(count)
# using numpy np.cumsum to calculate the CDF (cumulative distribution function)
# We can also find using the PDF values by looping and adding
cdf = np.cumsum(pdf)
# plotting PDF and CDF
# bins make the X axis while PDF & CDF are the corresponding Y values.
plt.plot(bins_count[1:], pdf, color="red", label="PDF")
plt.plot(bins_count[1:], cdf, label="CDF")
plt.legend()
```

## Marginal Plot

Marginal distribution plots are small subplots above or to the right of a main plot, which show the distribution of data along only one dimension. Marginal distribution plot capabilities are built into various Plotly Express functions such as scatter and histogram. Plotly Express is the easy-to-use, high-level interface to Plotly, which operates on a variety of types of data and produces easy-to-style figures.

To illustrate this, I'm using the Global mean air temperature anomalies (GMTA) and global Carbon Dioxide concentration for the years 1880 to 2004. Concentration is measured in parts per million (ppm). I'm asumming temperature anomalies are measured in Degrees Celcius.

```python
# For this, I'm going to use pandas
import pandas as pd
# It is a lot of information so I stored it in my Github for simplicity.
!wget https://raw.githubusercontent.com/coyoacan/Statistics/refs/heads/main/Source-files/T-anomaly-Co2-1880-2004.csv
anomaly = pd.read_csv('T-anomaly-Co2-1880-2004.csv')
```

```python
# Lets see the resulting dataframe's head
anomaly.head(10)
```

```python
# Use Plotly's Marginal Plot function
import plotly.express as px
fig = px.scatter(anomaly, x="Co2 Concentration", y="Anomaly",
                 marginal_x="box", marginal_y="box",
                  title="Marginal Plot for Temperature Anomaly and CO2 Concentration")
fig.show()
```
